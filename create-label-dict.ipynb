```python
import os
import shutil
import pandas as pd
```

```python
# 原本資料集
train_dir = r'.\模型訓練資料\train'
# 新建資料集
dataset_dir = r'./dataset/'

f_list = os.listdir(train_dir)
word_name_list = [w.split('.')[0] for w in f_list]
# 檔案編號與字對照
word_idx_dict = {}
for w in word_name_list:
    word_idx_dict[w.split('_')[0]] = w.split('_')[1]

word_df = pd.DataFrame({
    'idx': list(word_idx_dict.keys()),
    'word': list(word_idx_dict.values())
})
# 800字建立label
word_dict = {}
for i, w in enumerate(word_df['word'].unique()):
    word_dict[w] = i

word_dict_df = pd.DataFrame({
    'word': list(word_dict.keys()),
    'label': list(word_dict.values())
})
# 合併及存檔
word_df = pd.merge(word_df, word_dict_df, how='left', on='word')
word_df.to_csv(f'{dataset_dir}/word_df.csv', index=False)
```

```python
# copy & rename 
for f in f_list:
    new_f = f"{f.split('_')[0]}.jpg"
    shutil.copy(f'{train_dir}/{f}', f'{dataset_dir}{new_f}')
```
